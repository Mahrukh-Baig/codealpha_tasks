# FIBONACCI GENERATOR

The Fibonacci series is a sequence where each number isthe sum of the two preceding numbers, defined by a mathematical recurrence relationship.

```python
def fib(n):
    seq=[]
    first_num=0
    sec_num=1
    for i in range(n):
        seq.append(first_num)
        first_num,sec_num=sec_num ,first_num + sec_num
    return seq

num = int(input("Enter the number of Fibonacci terms: "))
print("Fibonacci Sequence:", fib(num))
        
```

# Visulization Of Fibonacci Sequence

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.plot(range(num), fib(num), marker='o', color='b', linestyle='-', linewidth=1, markersize=8)
plt.title('Fibonacci Sequence', fontsize=14)
plt.xlabel('Index (n)', fontsize=12)
plt.ylabel('Fibonacci Number (Fₙ)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(range(num))  
plt.show()
```
